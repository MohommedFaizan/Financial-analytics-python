```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('/Financial Analytics data.csv')
data
```

```
data.columns
```

```
data=data.rename(columns={'S.No.':'s_No','Mar Cap - Crore':'Mar_Cap_Core','Sales Qtr - Crore':'Sales_QTR_Core','Unnamed: 4':'new_col'})
data
```

```
data.dtypes
```

```
data['Sales_quater_Core'] = data['Sales_QTR_Core'].combine_first(data['new_col'])
data.head()
```

```
data.head(10)
```

```
data=data.drop(columns={'Sales_QTR_Core','new_col'})
data.head()
```

```
data.isnull().any()
```

```
sns.distplot(data['Mar_Cap_Core'])
```

```
col=data['Mar_Cap_Core']
col=col.fillna(col.median(),inplace=True)
```

```
data.isnull().sum()
```

```
sns.distplot(data['Sales_quater_Core'])
```

```
col=data['Sales_quater_Core']
col=col.fillna(col.median(),inplace=True)
```

```
data.isnull().sum()
```

```
data.duplicated()
```

```
top_10_brands = data.nlargest(10, 'Mar_Cap_Core')

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_10_brands['Name'], top_10_brands['Mar_Cap_Core'], color='skyblue')

# Add labels and title
plt.xlabel('Brand')
plt.ylabel('Market Capital Core')
plt.title('Top 10 Brands by Market Capital Core')
plt.xticks(rotation=45, ha='right')
```

```
top_10_sales_brands = data.nlargest(10, 'Sales_quater_Core')

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_10_sales_brands['Name'], top_10_sales_brands['Sales_quater_Core'], color='salmon')

# Add labels and title
plt.xlabel('Brand')
plt.ylabel('Sales Quarter Core')
plt.title('Top 10 Brands by Sales Quarter Core')
plt.xticks(rotation=45, ha='right')

# Show plot
plt.tight_layout()
plt.show()
```

```
top_10_brands = data.nlargest(10, 'Mar_Cap_Core')

# Extracting data for top 10 brands
market_cap_core_top_10 = top_10_brands['Mar_Cap_Core']
sales_qtr_core_top_10 = top_10_brands['Sales_quater_Core']
brands_top_10 = top_10_brands['Name']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(market_cap_core_top_10, sales_qtr_core_top_10, color='blue', alpha=0.5)

# Add labels and title
plt.xlabel('Market Capital Core')
plt.ylabel('Sales Quarter Core')
plt.title('Comparison of Market Capital Core and Sales Quarter Core of Top 10 Brands')

# Add brand names as annotations
for i, brand in enumerate(brands_top_10):
    plt.annotate(brand, (market_cap_core_top_10.iloc[i], sales_qtr_core_top_10.iloc[i]))

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
data['market_to_sales_ratio'] = data['Mar_Cap_Core'] / data['Sales_quater_Core']

# Sort the DataFrame by the market-to-sales ratio in ascending order (lower values indicate underperforming brands)
underperforming_brands = data.sort_values(by='market_to_sales_ratio', ascending=True)

# Select the top 10 underperforming brands
top_10_underperforming_brands = underperforming_brands.head(10)

# Display the top 10 underperforming brands
print(top_10_underperforming_brands[['Name', 'Mar_Cap_Core', 'Sales_quater_Core', 'market_to_sales_ratio']])

```

```
top_10_lowest_sales_brands = data.nsmallest(10, 'Sales_quater_Core')

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_10_lowest_sales_brands['Name'], top_10_lowest_sales_brands['Sales_quater_Core'], color='orange')

# Add labels and title
plt.xlabel('Sales Quarter Core')
plt.ylabel('Brand')
plt.title('Top 10 Brands with Lowest Sales Quarter Core')

# Show plot
plt.tight_layout()
plt.show()
```
